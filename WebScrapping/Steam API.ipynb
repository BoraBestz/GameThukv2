```python
def gameSpecMinimum_steam(steamId):
    response = requests.get("https://store.steampowered.com/api/appdetails?appids="+steamId+"&l=thai")
    gameJson = response.json()
    try:
        if (gameJson[steamId]['success'] == True) :
            gameDetail = gameJson[steamId]['data']
            gameSpec = gameDetail['pc_requirements']['minimum']
    except:
        print(steamId)
        gameSpec = ""
    return gameSpec
```

```python
def gameSpecRecommend_steam(steamId):
    response = requests.get("https://store.steampowered.com/api/appdetails?appids="+steamId+"&l=thai")
    gameJson = response.json()
    try:
        if (gameJson[steamId]['success'] == True) :
            gameDetail = gameJson[steamId]['data']
            gameSpec = gameDetail['pc_requirements']['recommended']
    except:
        print(steamId)
        gameSpec = ""
    return gameSpec
```

```python
def price_steam(steamId):
    response = requests.get("https://store.steampowered.com/api/appdetails?appids="+steamId+"&l=thai")
    gameJson = response.json()
    price = 0
    try :
        if (gameJson[steamId]['success'] == True) :
            gameDetail = gameJson[steamId]['data']
            price = float(gameDetail['price_overview']['final_formatted'].replace("฿","").replace(",",""))
    except:
        print(steamId)
    return price
    
```

```python
def detail_steam(steamId):
    response = requests.get("https://store.steampowered.com/api/appdetails?appids="+steamId+"&l=thai")
    gameJson = response.json()
    try :
        if (gameJson[steamId]['success'] == True) :
            gameDetail = gameJson[steamId]['data']
            description = gameDetail['short_description']
            releaseDate = gameDetail['release_date']['date']
            developer = gameDetail['developers'][0]
            publisher = gameDetail['publishers'][0]
            requiredAge = gameDetail['required_age']
            gameDetail_arr = [releaseDate,developer,publisher,requiredAge,description]
    except:
        print(steamId)
    return gameDetail_arr 
```
